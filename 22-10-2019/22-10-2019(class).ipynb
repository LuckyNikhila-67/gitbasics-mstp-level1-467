## lineplot using list comprehension


```python
import matplotlib.pyplot as p
x=[x for x in range(1,11)]
print(x)
y=[x**(1/2) for x in range(1,11)]
print(y)
p.plot(x,y)
p.grid()
```

```python
# draw a plot for time and distance by using list comprehension
## distance=2*time
import matplotlib.pyplot as p
t=[i for i in range(1,11,2)]
print(t)
d=[i*2 for i in range(1,11,2)]
print(d)
p.bar(t,d)
p.xlabel('time')
p.ylabel('distance')
p.title('time and distance')
```

### Matplotlib ticks---------?>ticks is values within coordinates

```python
import matplotlib.pyplot as p
x=[5,10,15,20,25,30,35,40,45,50]
print(x)
y=[9,8,5,6,3,2,1,4,7,10]
print(y)
p.plot(x,y)
p.title('A graph with example to ticks')
p.xlabel('natural numbers')
p.ylabel('values with incremented of 5 upto 50')
p.show()
```

```python
import matplotlib.pyplot as p
import numpy as np
p.yticks(np.arange(5,51,5))
p.xticks(np.arange(1,11))
p.plot(x,y)
p.title('Matplot-Ticks')
p.xlabel('n-natural numbers')
p.ylabel('5-50 values with 5 increment')
p.show()
```

### Representing sine wave
A=sine(wt),---
A=amplitude,--
w=frequency,--
t=time

```python
import numpy as np
import matplotlib.pyplot as p
t=np.arange(1,10,0.1)
Amplitude=np.sin(t)
p.plot(t,Amplitude)
p.xlabel('time')
p.ylabel('amplitude')
p.title('Sine wave')
p.show()
```

```python
import pandas as pd
df=pd.read_csv('income.csv')
df
```

```python
import matplotlib.pyplot as p
years=df.columns[2::]
alaska_income=df.iloc[1,2:]
p.plot(years,alaska_income)
p.xlabel('years of alaska')
p.title('income of alaska 2005--2013')
p.ylabel('year income of alaska')
p.savefig('income of alaska 2005--2013.jpg')
p.show()
```
